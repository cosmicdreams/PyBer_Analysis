```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
import statistics
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])

plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', color="skyblue")
plt.errorbar(pyber_ride_df["Month"], pyber_ride_df["Avg. Fare ($USD)"], yerr=stdev, capsize=3, ecolor="black")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.yticks(np.arange(0, 51, step=5.0))
plt.show()
```
